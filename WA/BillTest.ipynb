```python
import fitz
```

```python
def extract_text_from_pdf(path):
    text = ""
    with fitz.open(path) as doc:
        for page in doc:
            text += page.get_text()
    return text

pdf_text = extract_text_from_pdf("Bills/1003.PL.pdf")
print(pdf_text)
```

```python
def extract_text_from_pdf_after_sentence(path, start_sentence):
    text = ""
    found_start = False
    with fitz.open(path) as doc:
        for page in doc:
            page_text = page.get_text()
            if not found_start:
                index = page_text.find(start_sentence)
                if index != -1:
                    found_start = True
                    # Start collecting from the first occurrence of the sentence
                    text += page_text[index:]
            else:
                text += page_text
    return text
```

```python
start_sentence = "AN ACT"
pdf_text = extract_text_from_pdf_after_sentence("Bills/1003.PL.pdf", start_sentence)
print(pdf_text)
```
